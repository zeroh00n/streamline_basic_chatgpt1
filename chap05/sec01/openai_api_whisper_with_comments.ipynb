```python
# ========================= 코드 해설 주석 시작 =========================
# 이 셀의 역할(요약):
#  - OpenAI SDK로 Whisper 혹은 오디오/텍스트 관련 API를 호출합니다.
#  - .env에서 API 키 등 환경 변수를 로드합니다.
#  - 파일 경로/디렉터리를 탐색하고 입출력을 처리합니다.
# 주의: 아래 주석은 원본 코드의 동작을 변경하지 않으며, 가독성을 위해 상단에만 추가되었습니다.
# 첫 줄 미리보기: 'from openai import OpenAI'
# ========================= 코드 해설 주석 끝 ===========================
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv('OPENAI_API_KEY')
client = OpenAI(api_key=api_key)
```

```python
# ========================= 코드 해설 주석 시작 =========================
# 이 셀의 역할(요약):
#  - 오디오(음성) 입력을 처리하고 Whisper(STT) 또는 관련 API를 사용합니다.
# 주의: 아래 주석은 원본 코드의 동작을 변경하지 않으며, 가독성을 위해 상단에만 추가되었습니다.
# 첫 줄 미리보기: "audio_file_path = '../audio/lsy_audio_2023_58s.mp3' # MP3 파일 경로 입력"
# ========================= 코드 해설 주석 끝 ===========================
audio_file_path = '../audio/lsy_audio_2023_58s.mp3' # MP3 파일 경로 입력

with open(audio_file_path, 'rb') as audio_file:
    transcription = client.audio.transcriptions.create(
        model="whisper-1",
        file=audio_file
    )

transcription
```

```python
# ========================= 코드 해설 주석 시작 =========================
# 이 셀의 역할(요약):
#  - 오디오(음성) 입력을 처리하고 Whisper(STT) 또는 관련 API를 사용합니다.
# 주의: 아래 주석은 원본 코드의 동작을 변경하지 않으며, 가독성을 위해 상단에만 추가되었습니다.
# 첫 줄 미리보기: "with open(audio_file_path, 'rb') as audio_file:"
# ========================= 코드 해설 주석 끝 ===========================
with open(audio_file_path, 'rb') as audio_file:
    transcription = client.audio.translations.create(
        model="whisper-1", 
        file=audio_file, 
    )

transcription
```
